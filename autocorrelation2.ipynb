```python
%pylab inline

from sklearn.metrics.pairwise import pairwise_distances
from beat_spectrum import *
```

```python
max_lag = 200
window_size = 256
vecs = [np.sin(np.linspace(i, 2*np.pi + i, 12) * 2)
        + np.sin(np.linspace(i, 2*np.pi + i, 12) * 4)
        for i in np.linspace(0, 2*np.pi, window_size + max_lag)]
# vecs = np.random.rand(window_size + max_lag, 12)
S_frame = beat_similarity(vecs)
```

```python
plt.imshow(S_frame, cmap='gray')
```

```python
import scipy
B_fft = scipy.signal.fftconvolve(S_frame, S_frame[::-1, ::-1], mode='full')
```

```python
B_fft.shape
```

```python
B_fft_part = B_fft[S_frame.shape[0]:, S_frame.shape[1]:]
# plt.imshow(B_fft_part)

B_fft_sum = np.asarray([B_fft_part.trace(i) for i in range(max_lag)])
B_fft_sum /= np.max(B_fft_sum)
plt.plot(B_fft_sum[:max_lag])
```

```python
d = beat_diag(S_frame, max_lag)
s = beat_spectrum(S_frame, max_lag)
f = beat_spectrum_trace(S_frame, max_lag)

fft_s = beat_spectrum_fft(S_frame, max_lag)

pairwise_distances(np.asarray([d,s,f, fft_s]))
```

```python
plt.plot(range(200), fft_s, range(200), s, range(200), d, range(200), f)
```
