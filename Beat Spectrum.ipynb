```python
from scikits.audiolab import play

def play_amp(y, amp=0.25):
    assert amp < 1.0 and amp >= 0.0
    y -= y.min()
    y *= 2./y.max()
    y -= 1.0
    y *= amp
    assert y.min() >= -amp and y.min() <= amp
    assert np.isclose(y.max(), amp)
    play(y, 44100)
```
