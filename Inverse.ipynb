```python
def pad_y(arr, y_width, front=1):
    arr = np.asarray(arr)
    return np.pad(arr, ((front, y_width - arr.size - front)), mode='constant', constant_values=(0.0,0))

def normed(vec):
    vec = np.asarray(vec)
    return (vec - vec.min()) / (vec.max() - vec.min())
```

```python
from beat_spectrum import beat_spectra_from_track, beat_fft, beat_coefs

spectra = np.asarray([x for x in beat_spectra_from_track(i=0)])
ffts = beat_fft(spectra)[:, 1:25]
```

```python
%pylab inline

from scipy.fftpack import dct, idct
```

```python
def spectra_to_dct(spectra):
    return np.asarray([dct(spec, norm='ortho') for spec in spectra])[:, 1:25]

def dct_to_spectra(dcts):
    return np.asarray([idct(pad_y(x, 200)) for x in dcts])

fig, (ax1, ax2) = plt.subplots(1,2)

ax1.imshow(spectra.T, aspect='auto')
dcts = spectra_to_dct(spectra)
ax2.imshow(dct_to_spectra(dcts).T, aspect='auto')
```

```python
def fft_to_spectra(ffts):
    return np.asarray([np.fft.irfft(pad_y(np.exp(x), 101), 400)[:200] for x in ffts])

fig, (ax1, ax2) = plt.subplots(1,2)

ax1.imshow(spectra.T, aspect='auto')
ax2.imshow(fft_to_spectra(ffts).T, aspect='auto')
```
